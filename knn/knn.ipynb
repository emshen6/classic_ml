```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_n.drop('Wine', axis=1), df_n['Wine'], test_size=0.2, random_state=42)
```

```python
import numpy as np
import pandas as pd

class KNNClassifier:
    def predict(X_train, y_train, X_test, k):
        y_pred = []
        for i in range(X_test.shape[0]):
            distances = np.sqrt(np.sum((X_train - X_test.iloc[i])**2, axis=1))
            sorted_indices = np.argsort(distances)
            k_nearest_labels = y_train.iloc[sorted_indices[:k]]
            most_common_label = k_nearest_labels.mode().iloc[0]
            y_pred.append(most_common_label)

        return np.array(y_pred)
```

```python
num_features_to_select = np.random.randint(1, len(X_train.columns))  # Выбираем от 1 до (количество признаков - 1) признаков
X_train_selected_features = X_train.sample(n=num_features_to_select, axis=1)
X_test_selected_features = X_test[X_train_selected_features.columns]
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
k_value = [3, 5, 10, 100]
```

```python
for k in k_value:
  model1_predictions = KNNClassifier.predict(X_train_selected_features, y_train, X_test_selected_features, k)

  accuracy = accuracy_score(y_test, model1_predictions)
  print(f"Точность модели (k={k}): {accuracy * 100:.2f}%")

  # Строим матрицу ошибок
  cm = confusion_matrix(y_test, model1_predictions)
  print(cm)
  print()
```

```python
for k in k_value:
  model2_predictions = KNNClassifier.predict(X_train, y_train, X_test, k)

  accuracy = accuracy_score(y_test, model2_predictions)
  print(f"Точность модели (k={k}): {accuracy * 100:.2f}%")

  # Строим матрицу ошибок
  cm = confusion_matrix(y_test, model2_predictions)
  print(cm)
```
